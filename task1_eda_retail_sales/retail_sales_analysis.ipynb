```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("retail_sales.csv")

# Show first 5 rows
print(df.head())
```

```python
# Dataset information
print(df.info())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Statistical summary
print(df.describe())
```

```python
df['Age'].plot(kind='hist')

plt.title("Age Distribution")
plt.xlabel("Age")

plt.show()
```

```python
print(df.columns)
```

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Gender Distribution")

plt.ylabel('')

plt.show()
```

```python
top_spending = df['AmountSpent'].head(10)

top_spending.plot(kind='bar')

plt.title("Top 10 Customer Spending")
plt.xlabel("Customers")
plt.ylabel("Amount Spent")

plt.show()
```

```python
print(df.columns)
```

```python
top_spending = df['Total Amount'].head(10)

top_spending.plot(kind='bar')

plt.title("Top 10 Customer Spending")
plt.xlabel("Customers")
plt.ylabel("Total Amount")

plt.show()
```
